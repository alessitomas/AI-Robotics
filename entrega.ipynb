# Gráfico Objetivo x Tempo dos Algoritmos

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
df_profunidade = pd.read_csv("profunidades.csv", sep=",", encoding="latin1")
df_profunidade_iterativa = pd.read_csv("profunidade_iterativa.csv", sep=",", encoding="latin1")
df_busca_largura = pd.read_csv("algoritmos_micro.csv", sep=",", encoding="latin1")
df_profunidade['Tempo de processamento'] = df_profunidade['Tempo de processamento'].replace("Sem resultado",0)
df_profunidade['Tempo de processamento'] = [ int(i) for i in list(df_profunidade['Tempo de processamento'])]
df_profunidade_10 = df_profunidade[:50]
df_profunidade_100 = df_profunidade[50:]
```

```python

plt.figure(figsize=(15,8),dpi=1000)
sns.set(style="darkgrid")
sns.lineplot(x="Objetivo", y="Tempo de processamento", data=df_profunidade_10, label="m = 10")
sns.lineplot(x="Objetivo", y="Tempo de processamento", data=df_profunidade_100, label="m = 100")
sns.lineplot(x="Objetivo", y="Tempo de processamento", data=df_profunidade_iterativa, label="Busca em Profundidade Iterativa") 
sns.lineplot(x="Objetivo", y="Tempo de processamento", data=df_busca_largura, label="Busca em Largura")
plt.title("Tempo de processamento dos algoritmos de busca",fontsize=20)
plt.xlabel("Objetivo",fontsize=15)
plt.ylabel("Tempo de processamento (microsegundos)",fontsize=15)
# Configura o formato dos rótulos do eixo y para exibir os valores completos
plt.ticklabel_format(style='plain', axis='y')

plt.show()
```
