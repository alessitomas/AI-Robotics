# Busca Largura

```python
from algorithm import algoritmo

Algoritmo = ["Busca em Largura" for i in range(1,10)]
Objetivo = []
Tempo_de_processamento = []
for objetivo in range(1,51):
    print(objetivo)
    resultado, tempo = algoritmo(objetivo)
    Objetivo.append(objetivo)
    if resultado:
        Tempo_de_processamento.append(tempo)
    else:
        Tempo_de_processamento.append("Sem Resultado")

```

```python
Algoritmo = ["Busca em Largura" for i in range(1,51)]
```

```python
Tempo_de_processamento = [list(i)[0].microseconds for i in Tempo_de_processamento]
```

```python
import pandas as pd
dataframe = pd.DataFrame({"Algoritmo": Algoritmo,"Objetivo":Objetivo, "Tempo de processamento": Tempo_de_processamento})
```

```python
dataframe.head(50)
dataframe.to_csv("algoritmos_micro.csv")
```

# Busca Profunidade

```python
from algorithm import algoritmo_profunidade

Algoritmo_profunidade = []
Objetivo_profundidade = []
Tempo_de_processamento_busca_profundidade = []
profundidades = [10,100]

for m in profundidades:
    for objetivo in range(1,51):

        Algoritmo_profunidade.append(f"Busca em Profundidade com m = {m}")
        resultado, tempo = algoritmo_profunidade(objetivo,m)
        Objetivo_profundidade.append(objetivo)
        if resultado:
            Tempo_de_processamento_busca_profundidade.append(tempo)
        else:
            Tempo_de_processamento_busca_profundidade.append("Sem Resultado")
```

```python
Tempo_de_processamento_busca_profundidade = [list(i)[0].microseconds if type(list(i)[0]) != str else "Sem resultado"  for i in Tempo_de_processamento_busca_profundidade]
```

```python
import pandas as pd
dataframe_profunidade = pd.DataFrame({"Algoritmo": Algoritmo_profunidade,"Objetivo":Objetivo_profundidade, "Tempo de processamento": Tempo_de_processamento_busca_profundidade})
```

```python
dataframe_profunidade.to_csv("profunidades.csv", index=False)
```

# Busca Profunidade Iterativa

```python
from algorithm import algoritmo_profunidade

Algoritmo_profunidade_iterativa = ["Busca em Profundidade Iterativa" for i in range(1,51)]
Objetivo_profundidade_iterativa = []
Tempo_de_processamento_busca_profundidade_iterativa = []
for objetivo in range(1,51):
    print(objetivo)
    resultado, tempo = algoritmo_profunidade(objetivo)
    Objetivo_profundidade_iterativa.append(objetivo)
    if resultado:
        Tempo_de_processamento_busca_profundidade_iterativa.append(tempo)
    else:
        Tempo_de_processamento_busca_profundidade_iterativa.append("Sem Resultado")
```

```python
Tempo_de_processamento_busca_profundidade_iterativa = [list(i)[0].microseconds if type(list(i)[0]) != str else "Sem resultado"  for i in Tempo_de_processamento_busca_profundidade_iterativa]
```

```python
import pandas as pd
dataframe_profunidade_iterativa = pd.DataFrame({"Algoritmo": Algoritmo_profunidade_iterativa,"Objetivo":Objetivo_profundidade_iterativa, "Tempo de processamento": Tempo_de_processamento_busca_profundidade_iterativa})
```
